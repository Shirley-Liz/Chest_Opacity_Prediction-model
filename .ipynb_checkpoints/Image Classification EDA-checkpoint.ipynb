### Viewing a Sample of Images

```python

def list_sample_images(folder_path, num_samples):
    # get list of image file names
    file_names = os.listdir(folder_path)
    image_files = [i for i in file_names if i.endswith('.jpg') or i.endswith('.png')]

    # select random subset of images
    sample_files = random.sample(image_files, num_samples)

    # plot the selected images
    fig, ax = plt.subplots(1, num_samples, figsize=(10, 5))
    for j, file in enumerate(sample_files):
        img = imread(os.path.join(folder_path, file))
        ax[j].imshow(img)
        ax[j].axis('off')
    plt.show()
```

## Building the model

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

-Loading the dataset

```python

train_dir = "dataset/train/"
validation_dir = "dataset/validation/"
test_dir = "dataset/test/"

train_datagen = ImageDataGenerator(rescale = 1./255,
                               shear_range = 0.2,
                               zoom_range = 0.2,
                               vertical_flip = True)

validation_datagen = ImageDataGenerator(rescale=1./255)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

validation_generator = validation_datagen.flow_from_directory(
        validation_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='binary')
```

- The Image dataset has been split into 
- 80% Training set.
- 10% Validation set.
- 10% Testing set.

### Let's build the CNN model:

```python
model = tf.keras.models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

- Let's compile the model and train it:

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(
      train_generator,
      steps_per_epoch=train_generator.samples/train_generator.batch_size,
      epochs=20,
      validation_data=validation_generator,
      validation_steps=validation_generator.samples/validation_generator.batch_size)
```

### Model Loss & Accuracy

```python
fig = plt.figure()
plt.plot(history.history['loss'], color = 'teal', label = 'loss')
plt.plot(history.history['val_loss'], color = 'orange', label = 'val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(history.history['accuracy'], color = 'teal', label = 'accuracy')
plt.plot(history.history['val_accuracy'], color = 'orange', label = 'val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

### Model evaluation on the test set

```python
test_loss, test_acc = model.evaluate(test_generator, steps=test_generator.samples/test_generator.batch_size)
print('Test Accuracy:', test_acc)
```

- Checkout the labels

```python
labels = (train_generator.class_indices)
labels = dict((i,j) for j,i in labels.items())
print(labels)
```

### Classification Report and Confusion Matrix
- Predict on Test Set

```python
test_generator.reset()
predictions_test = (model.predict(test_generator,verbose=True) > 0.5).astype("int32")
```

```python
from sklearn.metrics import classification_report,confusion_matrix
y_true = np.array([0] * 36 + [1] * 63) #36 Normal, #63 Sick
data = confusion_matrix(y_true, predictions_test)
df_cm = pd.DataFrame(data, columns=np.unique(y_true), index = np.unique(y_true))
df_cm.index.name = 'Predicted'
df_cm.columns.name = 'Actual'
plt.figure(figsize = (10,7))
sns.set(font_scale=1.5)
ax = sns.heatmap(df_cm,cmap = 'Reds', annot=True,fmt = '.5g',annot_kws={"size": 16})# font size
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['NORMAL', 'SICK']); ax.yaxis.set_ticklabels(['NORMAL', 'SICK']);

print("Classification Report: ")
print(classification_report(y_true,predictions_test))
```

### Hyper parameter tuning

```python
hyper_dimension = 400
hyper_epochs = 100
hyper_batch_size = 16
hyper_feature_maps = 32
hyper_channels = 1
hyper_mode = 'grayscale'
```

```python
# Initializing the CNN
classifier = tf.keras.models.Sequential()

# Convolution & pooling - First convolution layer
classifier.add(layers.Conv2D(hyper_feature_maps, (3, 3),
                      input_shape = (hyper_dimension,
                                     hyper_dimension,
                                     hyper_channels),
                      activation = 'relu'))
classifier.add(layers.MaxPooling2D(pool_size = (2, 2)))

# Convolution & pooling - Second convolution layer (same as first layer)
classifier.add(layers.Conv2D(hyper_feature_maps, (3, 3),
                      input_shape = (hyper_dimension,
                                     hyper_dimension,
                                     hyper_channels),
                      activation = 'relu'))
classifier.add(layers.MaxPooling2D(pool_size = (2, 2)))

# Convolution & pooling - Third convolution layer
classifier.add(layers.Conv2D(hyper_feature_maps * 2, (3, 3),
                      input_shape = (hyper_dimension,
                                     hyper_dimension,
                                     hyper_channels),
                      activation = 'relu'))
classifier.add(layers.MaxPooling2D(pool_size = (2, 2)))

# Flattening
classifier.add(layers.Flatten())

# Full connection
classifier.add(layers.Dense(units = hyper_feature_maps * 2, activation = 'relu'))
classifier.add(layers.Dense(units = 1, activation = 'sigmoid'))

# Compiling the CNN
classifier.compile(optimizer = 'adam',
                   loss = 'binary_crossentropy',
                   metrics = ['accuracy'])
```

```python

# Creating training and validation image flows

train_gen = ImageDataGenerator(rescale = 1./255,
                               shear_range = 0.2,
                               zoom_range = 0.2,
                               vertical_flip = True)

val_gen = ImageDataGenerator(rescale = 1./255)

train_set = train_gen.flow_from_directory('dataset/train',
                                          target_size = (hyper_dimension,
                                                         hyper_dimension),
                                          batch_size = hyper_batch_size,
                                          class_mode = 'binary',
                                          color_mode = hyper_mode)

val_set = val_gen.flow_from_directory('dataset/validation',
                                      target_size = (hyper_dimension,
                                                     hyper_dimension),
                                      batch_size = hyper_batch_size,
                                      class_mode = 'binary',
                                      color_mode = hyper_mode)
```

```python
#Model Summary
classifier.summary()
```

```python
# Fitting the model

results = classifier.fit(train_set,
                                   steps_per_epoch = len(train_set),
                                   epochs = hyper_epochs,
                                   validation_data = val_set,
                                   validation_steps = len(val_set))
```

## Discussion and Final Remarks
Results show that very low losses and high accuracies (exceeding 96% in validation) are obtained with a traditional CNN approach, without any refinement based, for example, on the use of pretrained image classification models.

## Classify images in unknown folder
- Classify images in unknown folder using CNN Model 1

```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model

# Load the model
model = load_model('models/cnn_model_1.h5')

# Set the path to the folder containing the images to classify
image_folder = 'documents/dataset1/unknown_images'

# Get the list of image filenames in the folder
image_filenames = os.listdir(image_folder)

# Create an empty DataFrame to store the predictions
predictions_df = pd.DataFrame(columns=['Filename', 'Prediction'])

# Loop over each image in the folder
for filename in image_filenames:
    # Load the image and preprocess it
    image = load_img(os.path.join(image_folder, filename), target_size=(224, 224))
    image_array = img_to_array(image)
    image_array = np.expand_dims(image_array, axis=0)
    image_array = image_array / 255.0

    # Make the prediction
    prediction = model.predict(image_array)[0][0]

    # Add the prediction to the DataFrame
    df_new_row = pd.DataFrame({ 'Filename': [filename], 'Prediction': [round(prediction)] })
    predictions_df = pd.concat([predictions_df, df_new_row])
# Save the predictions to a CSV file
predictions_df.to_csv('results/unknown_images_predictions_model_1.csv', index=False)
```

### Test model performance on Dataset 2

```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.metrics import confusion_matrix, classification_report

# Load the model
model = tf.keras.models.load_model('models/cnn_model_1.h5')

# Set the path to the folder containing the test images
test_folder = 'doc/dataset2/images'

# Get the list of image filenames in the test folder
test_filenames = os.listdir(test_folder)

# Load the labels from the CSV file
labels_df = pd.read_csv('doc/dataset2/image_classes.csv')

# Create empty lists to store the true and predicted labels
true_labels = []
predicted_labels = []

# Loop over each test image and make a prediction
for filename in test_filenames:
    # Load the image and preprocess it
    image = load_img(os.path.join(test_folder, filename), target_size=(224, 224))
    image_array = img_to_array(image)
    image_array = np.expand_dims(image_array, axis=0)
    image_array = image_array / 255.0

    # Make the prediction
    prediction = model.predict(image_array)[0][0]
    predicted_label = int(round(prediction))
   
    #remove file extension because images in csv have no extension
    filename_no_extension = os.path.splitext(filename)[0]
    # Get the true label from the labels DataFrame
    true_label = labels_df.loc[labels_df['patient'] == filename_no_extension, 'label'].values[0]

    # Append the true and predicted labels to the lists
    true_labels.append(true_label)
    predicted_labels.append(predicted_label)
```

```python
# Print the classification report and confusion matrix
report = classification_report(true_labels, predicted_labels)
print(report)
# Save the text to a file
with open('results/dataset2_test_results.txt', 'w') as file:
    file.write(report)
# print(confusion_matrix(true_labels, predicted_labels))
```
